```python
import numpy as np
def get_vectors(path):
    vectors = dict()
    with open(path, 'r') as fhandle:
        for i, line in enumerate(fhandle):
            elements = line.split()
            if len(elements) > 2:
                try:
                    word = elements[0].lower()
                    vector = np.asarray([float(i) for i in elements[1:]])
                    vectors[word] = vector
                except:
                    pass
#                     print("Could not process line {}".format(i))
    return vectors
```

```python
def get_vectors_for_vocab(path, vocab):
    vectors = dict()
    with open(path, 'r', encoding='utf-8') as fhandle:
        for i, line in enumerate(fhandle):
            elements = line.split()
            if len(elements) > 2:
                try:
                    word = elements[0].lower()
                    if word in vocab:
                        vector = np.asarray([float(i) for i in elements[1:]])
                        vectors[word] = vector
                except:
                    pass
#                     print("Could not process line {}".format(i))
    return vectors
```

```python
danish_vectors = get_vectors_for_vocab('./data/embeddings/wiki.da.align.vec', vocab)
```

```python
def get_vocab_for_lang(lang):
    filename = './data/analogy_qids/analogy_all_{}.csv'.format(lang)
    entities = get_entities(filename)
    vocab = set()
    for entity in entities:
        es = entity.lower().split()
        for e in es:
            vocab.add(e)
    return vocab

filename = './data/embeddings/wiki.{}.align.vec'
for lang in languages:
    print("Language: {}".format(lang))
    vocab = get_vocab_for_lang(lang)
    print("Vocab: {}".format(len(vocab)))
    vectors = {l: get_vectors_for_vocab(filename.format(l), vocab) for l in tqdm(languages)}
    full_vocab_of_vectors = set()
    for l, v in vectors.items():
        full_vocab_of_vectors.update(v.keys())
    print("Language vectors: {}".format(len(vectors[lang])))
    print("Language augmented vectors: {}".format(len(full_vocab_of_vectors)))
    print("OOV: {}".format(1 - len(full_vocab_of_vectors) / len(vocab)))
```
